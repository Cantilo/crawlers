```python
# Импортируем необходимые библиотеки:
import requests # http-запросы,
import re # регулярные выражения,
from bs4 import BeautifulSoup # удаление тегов html,
import time
import random
import os
import tqdm
```

```python
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_2) AppleWebKit/600.3.18 (KHTML, like Gecko) Version/8.0.3 Safari/600.3.18'
headers = { 'User-Agent' : user_agent }
```

```python
startlink = r'https://panorama.pub/category/news/page/'
hreflist = []
```

```python
def GetLinks(startlink, rang=84):
    lst = []
    for letter in range(1,rang):
        link = startlink +str(letter)+'/'
        lst.append(link)
    return lst
```

```python
links = GetLinks(startlink)
```

```python
def GetTextLinks(link):
    lst = []
    r = requests.get(link,headers=headers)
    links = re.split("<h2 class=", r.text)
    for l in links[1:-1]:
        newlink = re.split('" rel="bookmark"',re.split('"entry-title"><a href="', l)[1])[0]
        lst.append(newlink)
    return lst
```

```python
listoftexts = []
for l in links:
    listoftexts+=GetTextLinks(l)
```

```python
stst_links =  GetLinks('https://panorama.pub/category/stati/page/',4)
```

```python
listofstats = []
for l in stst_links:
    listofstats+=GetTextLinks(l)
```

```python
dicofl = {}
for l in listoftexts:
    dicofl[l] = 'news'
for l in listofstats:
    dicofl[l] = 'article'
```

```python
alllinks = listoftexts+listofstats
```

```python
def GetText(link):
    try:
        lst = []
        r = requests.get(link,headers=headers)
        text = re.split('<div class="at-below-post addthis_tool',re.split('</div><p>', r.text)[1])[0]
        text = re.sub('&nbsp;', ' ', text)
        text = re.sub('&mdash;', ' -- ', text)
        text = re.sub('&shy;', '', text)
        text = re.sub('&laquo;', '"', text)
        text = re.sub('&raquo;', '"', text)
        beaux_text=BeautifulSoup(text, "lxml")
        text=beaux_text.get_text() 
        text = re.sub('\xa0', ' ', text)
        text = re.sub('\u2009', ' ', text)
        title = re.split('" />',re.split('<meta property="og:title" content="', r.text)[1])[0] #
        date = re.split('" />',re.split('<meta itemprop="datePublished" content="', r.text)[1])[0] #
        author = re.split('" />',re.split('<meta itemprop="author" content="', r.text)[1])[0] #
        return text, title, date, author
    except:
        print(link)
        return [],[],[],[]
```

```python
dic = {}
for l in tqdm.tqdm(alllinks):
    text_info = {}
    text_info['text'], text_info['title'], text_info['date'], text_info['author'] = GetText(l)
    text_info['link']=l
    text_info['type'] = dicofl[l]
    dic[l]=text_info
```

```python
def writetext(wdir, tid, text):
    try:
        f = open(os.path.join(wdir,str(tid)+'.txt'),'w', encoding='utf8')
        f.write(text)
        f.close
        return os.path.join(wdir,str(tid)+'.txt')
    except:
        print(text)
```

```python
metaname = r"D:\Panorama\metatable.csv"
foldername = r"D:\Panorama\texts"
metatable = open(metaname, "w", encoding="utf8")
order = ['link', 'title', 'date', 'author', 'type']
line = 'text_id'+ '\t' + '\t'.join(order)+'\t' +'filename'+"\n"
metatable.write(line)
idcount = 0
for l in tqdm.tqdm(dic):
    try:
        filename = writetext(foldername, idcount, dic[l]['text'])
        metatable.write(str(idcount)+'\t'+'\t'.join([dic[l][k] for k in order])+'\t'+filename+'\n')                                    
        idcount +=1
    except:
        pass

metatable.close()
```
