```python
def getHrefs(link):
    lst = []
    # Получаем текст страницы, которая содержит ссылки на все статьи этого дня (в примере - 11.01.2016).
    r = requests.get(link,headers=headers)
    
    # Каждая ссылка на статью оформлена с помощью тега <article>

    while len(lst)<37:
        refs = re.split("<a class=' ga-link' data-ga-context='talks' ", r.text)
        for i in refs:
            if "' lang='hy'>" in i:
                ref = re.split("href='", re.split("' lang='hy'>", i)[0])[1]
                ref = "https://www.ted.com" + ref

                lst.append(ref)

    for i in lst:
        hreffile.write(i+"\n")
        #print(i)


```

```python
link = "https://www.ted.com/talks?language=hy"
getHrefs(link)
for i in range(1,10):
    link2 = link+"&page="+str(i) 
    getHrefs(link2)
#https://www.ted.com/talks?language=hy&page=9"
```

```python
def newLink(link):
    filename_arm = "/media/mi_air/0F0B7DDE62EEA81E/Documents/TED/" + re.split("/", re.split("\?", link)[0])[4] +"_arm" + ".txt"
    file = open(filename_arm, "w", encoding="utf-8")
    link_arm = re.split("\?", link)[0] + "/transcript?language=hy"
    lst = getText(link_arm)
    for i in lst:
        file.write(i+"\n")
    filename_rus = "/media/mi_air/0F0B7DDE62EEA81E/Documents/TED/" + re.split("/", re.split("\?", link)[0])[4] +"_rus" + ".txt"
    file2 = open(filename_rus, "w", encoding="utf-8")
    link_rus = re.split("\?", link)[0] + "/transcript?language=ru"
    lst2 = getText(link_rus)
    for i in lst2:
        file2.write(i+"\n")
```

```python
def getText(link):
    text_lst =[]
    r = requests.get(link,headers=headers)
    lst = re.split("<span class='talk-transcript__para__text'>", r.text)
    
    for i in lst[1:]:
        
        text = ""
        a = []
        txt =re.split("<span class='talk-transcript__fragment'", i)
        
        for j in txt:
            if ">" in j:
                line = re.split(">", re.split("</span>",j)[0])[1]
                if "&quot;" in line:
                    line = re.sub("&quot;", "'", line)
                a.append(line.strip("\n"))
        text = " ".join(a)
        text_lst.append(text)
    return(text_lst)
```

```python
for item in lst:
    newLink(item)
    print(item)
```
