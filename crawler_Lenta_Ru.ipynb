```python
# Импортируем необходимые библиотеки:
import requests # http-запросы,
import re # регулярные выражения,
from bs4 import BeautifulSoup # удаление тегов html,
from tqdm import tqdm # красотуля для анализа прогресса.
import time
import random
```

```python
def getHrefs(link):

    # Получаем текст страницы, которая содержит ссылки на все статьи этого дня (в примере - 11.01.2016).
    r = requests.get(link)
    # Каждая ссылка на статью оформлена с помощью тега <article>
    refs=re.split('href="', re.split('<svg class="control_mini-icon control_mini-icon-prev">', re.split('">Все материалы</a></div></div></div><div id="context_3"></div></div></section><div id="root_footer">', r.text)[0])[1])    
    # Если ссылки вообще есть, ...
    for i in refs:
        if i.startswith("/news/"):
            ilink = re.split('">',i)[0]
            hreffile.write("https://lenta.ru/"+ilink+ "\n")
        if i.startswith("/article/"):
            ilink = re.split('">',i)[0]
            #print(ilink)
            hreffile.write("https://lenta.ru/"+ilink+ "\n")

            #time.sleep(random.uniform(1, 4))
        
```
