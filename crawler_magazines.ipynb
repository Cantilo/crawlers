```python
# Импортируем необходимые библиотеки:
import requests # http-запросы,
import re # регулярные выражения,
from bs4 import BeautifulSoup # удаление тегов html,
from tqdm import tqdm # красотуля для анализа прогресса.
import time
import random
```

```python
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_2) AppleWebKit/600.3.18 (KHTML, like Gecko) Version/8.0.3 Safari/600.3.18'
headers = { 'User-Agent' : user_agent }
```

```python

hreffile=open(r"/home/mi_air/Downloads/Magazines/href_list_dedup.txt", "w", encoding="utf-8")
hrefnumero=open(r"/home/mi_air/Downloads/Magazines/href_numero.txt", "w", encoding="utf-8")
```

```python
#есть список журналов, у них- - список номеров и список всех статей в номерах
#список журнала взяли с главной, теперь важно собрать все ссылки на номера

def getNumero(journallink):
    journ = journallink.split('/')[3]
    r = requests.get(journallink,headers=headers)
    links = re.split('<td><a href', re.split('</table>', re.split('<p>Архив:</p>', r.text)[1])[0])
    for line in links[1:]:
        if '="' in line:
            numerolink = re.split('">', re.split('="',line)[1])[0]
            print(numerolink)
            hrefnumero.write("http://magazines.russ.ru" + numerolink+"\n")
```

```python
def GetArticles(link):
    r = requests.get(link,headers=headers)
    links = re.split('<strong>', r.text)
    for items in links[1:]:
        if "<a href=" in items:
            textlink = re.split('">', re.split('<a href="', items)[1])[0]
            #print(textlink)
            hreffile.write("http://magazines.russ.ru" + textlink+"\n")
```

```python
numeros = open(r"/home/mi_air/Downloads/Magazines/href_numero.txt", "r", encoding="utf-8")
for line in numeros:
    link = line.strip("\n")
    GetArticles(link)
```

```python
def GetTextMag(textlink):
    r = requests.get(textlink,headers=headers)
    title = re.split('"/>', re.split(' <meta property="og:title" content="', r.text)[1])[0] 
    authorname = re.split('"/>', re.split('<meta property="article:author" content="', r.text)[1])[0]
    textid = re.split('"/>', re.split('<meta property="article:id" content="', r.text)[1])[0]
    texttype = re.split('"/>', re.split('<meta property="og:description" content="', r.text)[1])[0]
    publ = re.split('<a href=', re.split('</p>', re.split('Опубликовано в журнале:', r.text)[1])[0])
    date = re.split('</a>', re.split('">', publ[2])[1])[0]
    magazine = re.split('</a>', re.split('">', publ[1])[1])[0]
    abz = re.split('<p class=MsoBodyText',r.text)
    text = ""
    for p in abz[1:]:
        text = text + re.split('<o:p></o:p>', re.split("windowtext'>", p)[1])[0]
    
    beaux_text=BeautifulSoup(text, "lxml")
    n_text=beaux_text.get_text() 
    n_text = re.sub("\r\n", " ", n_text)
    n_text = re.sub("\xa0", " ", n_text)
    time.sleep(random.uniform(0, 2))
    return [title, authorname, textid, texttype, date, magazine, n_text]
```

```python
print(GetTextMag('http://magazines.russ.ru/znamia/2016/1/gnev.html'))
```

```python
corpus = open(r"/home/mi_air/Downloads/Magazines/corpus.txt", "w", encoding="utf-8")
hreffile=open(r"/home/mi_air/Downloads/Magazines/href_list_dedup.txt", "r", encoding="utf-8")
for line in hreffile:
    try:
        #[title, authorname, textid, texttype, date, magazine, n_text]
        params = GetTextMag(line.strip("\n"))
        corpus.write("<textname>"+params[0]+"</textname>"+"\n")
        corpus.write("<textrubric>"+params[3]+"</textrubric>"+"\n")
        corpus.write("<date>"+params[4]+"</date>"+"\n")
        corpus.write("<magazine>"+params[5]+"</magazine>"+"\n")
        corpus.write("<source>"+line.strip("\n")+"</source>"+"\n")
        corpus.write("<textid>" + params[2] +"</textid>" +"\n")
        corpus.write("<text>"+"\n")
        corpus.write(params[6]+"\n")
        corpus.write("</text>"+"\n")
        print(line.strip("\n"))
        
    except:
        pass
```
